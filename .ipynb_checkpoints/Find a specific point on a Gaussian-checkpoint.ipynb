```python
# y = e^( (-4(ln)(2)(t)^2)/(h^2) )
# portion in () is all in exponent


# also y = exp((-4(ln)(2)(t)^2)/(h^2))

# y = y coordinate
# t = x coordinate
# h = fullwidth at half maximum

# to solve for a given point, take natural log (ln)
#of both sides of equation
#(ie: ln(y) = ln(  e^( (-4(ln)(2)(t)^2)/(h^2) )  )
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# gaussian

h = .999 # this is the original parameter

t = np.linspace(-3,3,1001)

g = np.exp(-4*np.log(2)*t**2 / h**2)

plt.plot(t,g)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Gaussian with FWHM=' + str(h))
plt.show()
```

```python
yVal2find = .5
# mat = np.zeros((np.shape(t)[0],np.shape(t)[0]))
# yVal2find = list(range(len(t)))

tvals = h*np.sqrt( np.log(yVal2find)/(-4*np.log(2)))

print(tvals,-tvals)
```

```python
# exersise.independent # using parameters from cells above
shape = 1001
h = np.linspace(0,1,shape)

t = np.linspace(-3,3,shape)

for i in range(shape):
    print(np.exp(-4*np.log(2)*t[i]**2 / h[i]**2))
# g = np.exp(-4*np.log(2)*t**2 / h**2)
# print(t)
```

```python
# experimental
N = 100
mat = np.zeros((np.shape(t)[0],np.shape(t)[0]))
# mat = np.zeros()

for i in range(len(t)):
    for ci in range(len(t)):
        h = (i+1)/N
        mat[i,ci] = np.exp(-4*np.log(2)*t[i]**2 / h**2)
plt.pcolormesh(mat)
```

```python
# exercise.instructed

N = 100

G = np.zeros((N,len(t)))
h = np.zeros(N)

for i in range(N):
    h[i] = (i+1)/N
    G[i,:] = np.exp( -4*np.log(2)*t**2 / h[i]**2 )
    
plt.pcolormesh(G)
plt.xlabel('t')
plt.ylabel('h')
plt.show()
```
