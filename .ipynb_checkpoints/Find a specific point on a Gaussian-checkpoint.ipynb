```python
# y = e^( (-4(ln)(2)(t)^2)/(h^2) )
# portion in () is all in exponent


# also y = exp((-4(ln)(2)(t)^2)/(h^2))

# y = y coordinate
# t = x coordinate
# h = fullwidth at half maximum

# to solve for a given point, take natural log (ln)
#of both sides of equation
#(ie: ln(y) = ln(  e^( (-4(ln)(2)(t)^2)/(h^2) )  )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display,Math
import sympy as sym
```

```python
# gaussian

h = .999 # this is the original parameter

t = np.linspace(-3,3,1001)

g = np.exp(-4*np.log(2)*t**2 / h**2)


plt.plot(t,g)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Gaussian with FWHM=' + str(h))
plt.show()
```

```python

N = 100

G = np.zeros((N,len(t)))
h = np.zeros(N)

for i in range(N):
    h[i] = (i+1)/N
    G[i,:] = np.exp( -4*np.log(2)*t**2 / h[i]**2 )
    
plt.pcolormesh(G)
plt.xlabel('t')
plt.ylabel('h')
plt.show()
```
