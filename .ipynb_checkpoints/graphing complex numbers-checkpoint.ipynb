```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# complex numbers have two parts
# ie: z = 2 3i    2 = real part, 3i = imaginary part
```

```python
# parts in complex number can be thought of as
# coordinates on cart graph x = real, y = imag
```

```python
z = np.complex(2,3)

# plt.plot(z, 'ro') # discards imag part

plt.plot(np.real(z),np.imag(z), 'ro-')
# plt.plot(np.real(z),np.imag(z), '-')
plt.plot(3,4,3,6, 'r-')
plt.xlabel('real')
plt.ylabel('imaginary')
plt.grid()
plt.show()
plt.axis([-4,4,-4,4])
```

```python
z = np.complex(-3,-1)

# plt.plot(z, 'ro') # discards imag part

plt.plot(np.real(z),np.imag(z), 'ro')
plt.plot([0,np.real(z)],[0,np.imag(z)],'r')
# plt.plot(np.real(z),np.imag(z), '-')

plt.xlabel('real')
plt.ylabel('imaginary')
plt.grid()
plt.axis([-4,4,-4,4])
plt.show()
```

```python
# exercise.independent
z1 = np.complex(-2,2)
z2 = np.complex(-2,-1)
z3 = z1+z2 #sum
z4 = z1*z2 #product

ax_dems = [-5,5] #axis demensions

# sum plots:
plt.plot([0,np.real(z1)],[0,np.imag(z1)],'b',label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],'g',label='z2')
plt.plot([0,np.real(z3)],[0,np.imag(z3)],'r',label='z1+z2')

plt.legend()
plt.axis('square')
plt.axis([ax_dems[0],ax_dems[1],ax_dems[0],ax_dems[1]])
plt.grid()
plt.xlabel('Real')
plt.ylabel('Imag')
plt.subplots()

# product plots
plt.plot([0,np.real(z1)],[0,np.imag(z1)],'b',label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],'g',label='z2')
plt.plot([0,np.real(z4)],[0,np.imag(z4)],'r',label='z1 x z2')

plt.legend()
plt.axis('square')
plt.grid()
plt.xlabel('Real')
plt.ylabel('Imag')
plt.axis([ax_dems[0],ax_dems[1],ax_dems[0],ax_dems[1]])

plt.show()
```

```python
# exercise.instructed
z1 = np.complex(-3,1)
z2 = np.complex(-1,1)

z1pz2 = z1+z2

plt.plot([0,np.real(z1)],[0,np.imag(z1)],label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],label='z2')
plt.plot([0,np.real(z1pz2)],[0,np.imag(z1pz2)],label='z1+z2')

plt.xlabel('real')
plt.ylabel('imaginary')
plt.grid()
plt.axis('square')
plt.legend()
plt.axis([-5,5,-5,5])
plt.show()
```

```python
# part 2 of instructed exercise

z1tz2 = z1*z2

plt.plot([0,np.real(z1)],[0,np.imag(z1)],label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],label='z2')
plt.plot([0,np.real(z1tz2)],[0,np.imag(z1tz2)],label='z1*z2')

plt.xlabel('real')
plt.ylabel('imaginary')
plt.grid()
plt.axis('square')
plt.legend()
plt.axis([-5,5,-5,5])
plt.show()
```
