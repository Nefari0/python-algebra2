```python
# discreet derivative yt = xt + 1 - xt
```

```python
x = [1,5,4]
```

```python
np.diff(x) # compute difference between current point and next point
```

```python

x = np.linspace(-2,2,101)
f = x**2
df = np.diff(f)

plt.plot(x,f,label='f')
plt.plot(x[0:-1],df,label='df')

plt.grid()
plt.axis([-2,2,-1,2])
plt.legend()
plt.show()
```
