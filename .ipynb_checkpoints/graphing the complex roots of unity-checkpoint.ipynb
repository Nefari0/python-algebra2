```python
# graphing the complex roots of unity
#  z**n = 1, where z = some complex number

#  find z with formula:
# z = e**(2*pi*i*k/n)
# where k is set of numbers 0 to n-1
```

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
n = 17 # works with other numbers as well

# -----original with rounding error----
# for k in range(0,n):
#     z = np.exp( 2*np.pi*1j*k/n)
#     print(z**n)

# -----using sympy-----
for k in range(0,n):
    z = sym.exp( 2*sym.pi*sym.I*k/n)
    display(Math('(%s)^{%s} \\Rightarrow %s' %(sym.latex(z),n,sym.latex(z**n))))
```

```python
# plotting the roots
n = 15

for k in range(0,n):
    z = np.exp( 2*np.pi*1j*k/n)
    plt.plot([0,np.real(z)],[0,np.imag(z)])
    
plt.axis('square')

x = np.linspace(0,2*np.pi,100)
plt.plot(np.cos(x),np.sin(x),color='grey')
```

```python
# exercise.independent

n = 200

for k in range(0,n):
    z = k*np.exp( 2*np.pi*1j*k/n)
    plt.plot([0,np.real(z)],[0,np.imag(z)],color='grey')
    
plt.axis('square')

x = np.linspace(0,2*np.pi,100)
plt.plot(np.cos(x),np.sin(x),color='grey')
```

```python
# exercise.instructed

n = 200
color = np.linspace(0,.9,n)

for k in range(0,n):
    z = k*np.exp( 2*np.pi*1j*k/n)
    plt.plot([0,np.real(z)],[0,np.imag(z)],color=[color[k],color[k],color[k]])
    
plt.axis('square')
plt.axis('off')
plt.show()
    
```
